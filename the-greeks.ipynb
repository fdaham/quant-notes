# The Greeks

The greeks are first and second order partial derivatives of the option price with respect to different underlying factors such as the stock price, time, and interest. They're used to measure the risk or return on the financial derivative, $f$. 

## Delta

Delta has the greatest impact on the option's price (or premium). It:

* Measures how much the option's price changes when the underlying (stock) price changes by $\$$1 (e.g. if $\Delta = 0.3$, then the option's price is expected to increase by $\$$0.30).  
* Measures the probability of expiring in the money (ITM) (e.g. if $\Delta = 0.3$, then the option has a 30% chance of expiring ITM).
* Is not constant. 

\begin{equation*}
\Delta = \frac{\delta f}{\delta S}
\label{eq:1} \tag{1}
\end{equation*}

For European calls, $c$, and puts, $p$, with dividend yield $y$, delta equates to: 

\begin{equation*}
\begin{split}
\Delta_c &= e^{-y\tau}N(d_1) \\
\Delta_p &= -e^{-y\tau}[1-N(d_1)]
\end{split}
\label{eq:2} \tag{2}
\end{equation*}

## Gamma

Gamma measures delta's expected rate of change (e.g. if $\Delta = 0.3$ and $\Gamma = 0.05$, then the premium is expected to increase by $\$ 0.35$ on the next dollar move). 

\begin{equation*}
\Gamma = \frac{\delta^2 f}{\delta S^2}
\label{eq:3} \tag{3}
\end{equation*}

For European calls and puts, we have:

\begin{equation*}
\Gamma = \frac{N'(d_1)e^{-y\tau}}{S \sigma \sqrt{\tau}}
\label{eq:4} \tag{4}
\end{equation*}

## Theta

The greek of time decay. Theta measures the estimated value lost by an option with each passing day (e.g. if $\Theta = -0.04$, the option's price is expected to go down $\$0.04$ in value daily). Due to this, Theta works against buyers and for sellers. 

\begin{equation*}
\Theta = \frac{\delta f}{\delta t}
\label{eq:5} \tag{5}
\end{equation*}

For European calls and puts, we have:

\begin{equation*}
\begin{split}
\Theta_c &= - \frac{SN'(d_1)\sigma e^{-y \tau}}{2 \sqrt{\tau}} + ySe^{-y \tau}N(d_1) - rKe^{-y \tau}N(d_2) \\
\Theta_p &= - \frac{SN'(d_1)\sigma e^{-y \tau}}{2 \sqrt{\tau}} + ySe^{-y \tau}N(-d_1) + rKe^{-y \tau}N(-d_2)
\end{split}
\label{eq:6} \tag{6}
\end{equation*}

## Vega

The greek of implied volatility. Vega measures how much the premium changes with each $1\%$ change in implied volatility (e.g. if $\nu = 0.03$ and implied volatility decreases by $1\%$, then the option's price is expected to drop by $\$0.03$). The impact of implied volatility depends on the strategy used when pricing the option. Also, the further out the time to expiry is, the higher vega will be.    

\begin{equation*}
\nu = \frac{\delta f}{\delta r}
\label{eq:7} \tag{7}
\end{equation*}

For European options, we have:

\begin{equation*}
\begin{split}
\nu &= \frac{\delta c}{\delta \sigma} = \frac{\delta p}{\delta \sigma} 
&= Se^{-y\tau}\sqrt{\tau}N'(d_1)
\end{split}
\label{eq:8} \tag{8}
\end{equation*}
