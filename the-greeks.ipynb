# The Greeks

The greeks are first and second order partial derivatives of the option price with respect to different underlying factors such as the stock price, time, and interest. They're used to measure the risk or return on a financial derivative, $f$. 

## Delta

Delta has the greatest impact on the option's price (or premium). It measures how much the option's price changes when the underlying (stock) price changes by $\$$1 (e.g. if $\Delta = 0.3$, then the option's price is expected to increase by $\$$0.30). It also loosely equals the probability of finishing in the money (ITM, or $S > K$ for a call and $S < K$ for a put) (e.g. if $\Delta = 0.3$, then the option has a 30% chance of expiring ITM). Delta's value is constant. 

\begin{equation*}
\Delta = \frac{\delta f}{\delta S}
\label{eq:1} \tag{1}
\end{equation*}

For European calls, $c$, and puts, $p$, with dividend yield $y$, delta equates to: 

\begin{equation*}
\begin{split}
\Delta_c &= e^{-y\tau}N(d_1) \\
\Delta_p &= -e^{-y\tau}[1-N(d_1)]
\end{split}
\label{eq:2} \tag{2}
\end{equation*}

where $0 \leq \Delta_c \leq 1$ and $-1 \leq \Delta_p \leq 0$.

### Delta Hedging

Delta hedging is the process of keeping the delta of a portfolio (consisting of assets and options on those assets) close to zero. This is done by either buying or selling some of the assets such that the total exposure from the assets plus the options are zero. The goal is to make the future realized volatility (the historical annualized standard deviation of an asset’s log returns) higher than the implied volatility ($\Delta \sigma$, an asset’s assumed volatility over the option period)--which is considered a negative volatility spread. 

The daily PnL (Profit and Loss) of a delta-hedged position is equal to the sum of the Gamma, Theta, and Vega PnL's from that day:

\begin{equation*}
\text{Daily PnL} = \frac{1}{2}\Gamma (\Delta S)^2 + \Theta (\Delta t) + \nu (\Delta \sigma)
\label{eq:3} \tag{3}
\end{equation*}

## Gamma

Gamma measures delta's expected rate of change (e.g. if $\Delta = 0.3$ and $\Gamma = 0.05$, then the premium is expected to increase by $\$ 0.35$ on the next dollar move). In other words, it shows the option delta’s sensitivity to market price changes. Gamma reaches its maximum value when the option is at the money (ATM, or $S = K$).

\begin{equation*}
\Gamma = \frac{\delta^2 f}{\delta S^2}
\label{eq:4} \tag{4}
\end{equation*}

For European calls and puts, we have:

\begin{equation*}
\Gamma = \frac{N'(d_1)e^{-y\tau}}{S \sigma \sqrt{\tau}}
\label{eq:5} \tag{5}
\end{equation*}

## Theta

The greek of time decay. Theta measures the estimated value lost by an option with each passing day (e.g. if $\Theta = -0.04$, the option's price is expected to go down $\$0.04$ in value daily). Due to this, Theta works against buyers and for sellers. 

\begin{equation*}
\Theta = \frac{\delta f}{\delta t}
\label{eq:6} \tag{6}
\end{equation*}

For European calls and puts, we have:

\begin{equation*}
\begin{split}
\Theta_c &= - \frac{SN'(d_1)\sigma e^{-y \tau}}{2 \sqrt{\tau}} + ySe^{-y \tau}N(d_1) - rKe^{-y \tau}N(d_2) \\
\Theta_p &= - \frac{SN'(d_1)\sigma e^{-y \tau}}{2 \sqrt{\tau}} + ySe^{-y \tau}N(-d_1) + rKe^{-y \tau}N(-d_2)
\end{split}
\label{eq:7} \tag{7}
\end{equation*}

## Vega

The greek of implied volatility, $\sigma$. Vega measures how much the premium changes with each $1\%$ change in implied volatility (e.g. if $\nu = 0.03$ and implied volatility decreases by $1\%$, then the option's price is expected to drop by $\$0.03$). The further out the time to expiry is, the higher vega will be. 

The impact of implied volatility depends on the strategy used when pricing the option. Some Vega-sensitive strategies include straddles, where a profit can be made when volatility increases or decreases without a move in the underlying asset.

\begin{equation*}
\nu = \frac{\delta f}{\delta \sigma}
\label{eq:8} \tag{8}
\end{equation*}

For European options, we have:

\begin{equation*}
\begin{split}
\nu &= \frac{\delta f_c}{\delta \sigma} = \frac{\delta f_p}{\delta \sigma} 
&= Se^{-y\tau}\sqrt{\tau}N'(d_1)
\end{split}
\label{eq:9} \tag{9}
\end{equation*}

## Rho

Rho measures an option’s sensitivity to changes in the risk-free interest rate, $r$. Since interest rates change slowly over time, they have a smaller impact on the option's price. Rho is directly proportional to the expiration time; the longer the time to expiry, the higher the impact of the interest rates and, consequently, Rho. 

\begin{equation*}
\rho = \frac{\delta f}{\delta r}
\label{eq:10} \tag{10}
\end{equation*}

For European options, we have:

\begin{equation*}
\begin{split}
\rho_c &= Kte^{-rt}N(d_2) \\
\rho_p &= -Kte^{-rt}N(-d_2)
\end{split}
\label{eq:11} \tag{11}
\end{equation*}
