# Cauchy Distributions

Cauchy Distributions are continuous distributions that represent forced resonance. In spectroscopy, it's used to describe the shape of spectral lines. In addition to its usefulness in physics, the Cauchy distribution is used in finance to describe deviations in returns from predictive models. The PDF of a standard Cauchy distribution is:

\begin{equation*}
f(x) = \frac{1}{\pi(1 + x^2)}
\label{eq:1} \tag{1}
\end{equation*}

and the CDF is:

\begin{equation*}
F(x) = \frac{1}{2} + \frac{\text{arctan}(x)}{\pi}
\label{eq:2} \tag{2}
\end{equation*}

## Mean

The Cauchy distribution has no defined mean. Let's prove it:

\begin{equation*}
\begin{split}
E[X] &= \int_{- \infty}^{\infty} x f(x) dx \\
&= \frac{1}{\pi} \int_{- \infty}^{\infty} \frac{x}{1+x^2} dx
\end{split}
\label{eq:3} \tag{3}
\end{equation*}

Substituting $t = 1 + x^2 \Rightarrow dt = 2xdx$, we have:

\begin{equation*}
\begin{split}
E[X] &= \frac{1}{2\pi} \int_{- \infty}^{\infty} \frac{1}{t} dt \\
&= \frac{1}{2\pi} ln(t)  \Biggr]_{- \infty}^{\infty} = \frac{ln \Biggr( \frac{\infty}{- \infty} \Biggl)}{2 \pi}
\end{split}
\label{eq:4} \tag{4}
\end{equation*}

which is undefined.

## Variance

The variance of a Cauchy distribution is infinite. Although the first moment is undefined, the second moment does exist. Let's prove it:

\begin{equation*}
\begin{split}
E[X^2] &= \int_{- \infty}^{\infty} x^2 f(x) dx \\
&= \frac{1}{\pi} \int_{- \infty}^{\infty} \frac{x^2}{1+x^2} dx = \frac{1}{\pi} \int_{- \infty}^{\infty} \Biggl( 1 - \frac{1}{1+x^2} \Biggr) dx \\
&= \frac{1}{\pi} \int_{- \infty}^{\infty} dx - \frac{1}{\pi} \int_{- \infty}^{\infty} \frac{1}{1+x^2} dx
\end{split}
\label{eq:5} \tag{5}
\end{equation*}

where $\frac{1}{\pi} \int_{- \infty}^{\infty} \frac{1}{1+x^2} dx$ is the PDF of a standard Cauchy distribution with an integral of 1: 

\begin{equation*}
\begin{split}
\frac{1}{\pi} \int_{- \infty}^{\infty} \frac{1}{1 + x^2} dx &= \frac{1}{\pi} \Biggl[ \text{tan}^{-1}(x) \Biggr]_{- \infty}^{\infty} \\
&= \frac{1}{\pi} \Biggl[ \frac{\pi}{2} - (-\frac{\pi}{2}) \Biggr] = 1
\end{split}
\label{eq:6} \tag{6}
\end{equation*}

Therefore:

\begin{equation*}
E[X^2] = \frac{1}{\pi} x \Biggr]_{- \infty}^{\infty} - 1 = \frac{1}{\pi} [\infty + \infty] - 1 = \infty 
\label{eq:7} \tag{7}
\end{equation*}

And, since the mean is undefined, we have $\text{var}(X) = E[X^2] = \infty$.
