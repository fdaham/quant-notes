# Exponential and Poisson Distributions

## Exponential Distributions

An exponential distribution is used to model the time interval between independent events that happen at a constant average rate, $\lambda$. It has the following PDF with an expected arrival time of $\frac{1}{\lambda}$ and variance $\frac{1}{\lambda^2}$.

\begin{equation*}
f(t)=\begin{cases}
          \lambda e^{-\lambda t} \quad &t \geq 0 \\
          0 \quad &\text{otherwise} \\
     \end{cases}
\label{eq:1} \tag{1}
\end{equation*}

By integrating the PDF above, we can derive the CDF of an exponential distribution:

\begin{equation*}
F(t)=\begin{cases}
          P(\tau \leq t) = 1 - e^{-\lambda t} \\
          P(\tau > t) = e^{-\lambda t} \\
     \end{cases}
\label{eq:2} \tag{2}
\end{equation*}

where $\tau$ is the random variable for arrival time. Exponential distributions have a memorylessness property; the time to the next arrival is independent of the previous arrival. In other words, waiting for an additional time $s$ has the same distribution as waiting without any additional time.

\begin{equation*}
\begin{split}
P(\tau > s + t | \tau > s) &= \frac{e^{-\lambda (s+t)}}{e^{- \lambda s}} \\
&= e^{- \lambda t} \\
&= P(\tau > t)
\end{split}
\label{eq:3} \tag{3}
\end{equation*}

When a series of events independently follow an exponential distribution with an arrival rate of $\lambda$, the number of arrivals within the time interval $[0,t]$ can be modeled as a Poisson process. 

## Poisson Distributions

A Poisson distribution is used to predict the probability of an event happening in a fixed interval of time given how often the event occurs. The PMF of a Poisson distributed discrete random variable $X$ is:

\begin{equation*}
P(X) = \frac{e^{- \lambda} \lambda^x}{x!}
\label{eq:4} \tag{4}
\end{equation*}

where $x \in [0, \infty)$ is the observed number of events, $\lambda$ is the arrival rate, $E[X] = \lambda$, and $\text{var}(X) = \lambda$. See [here](https://llc.stat.purdue.edu/2014/41600/notes/prob1804.pdf) for the derivation of $E[X]$ and $\text{var}(X)$.

### Poisson Process

As mentioned above, a Poisson process is a statistical process where the number of events occurring within an independent time interval is modeled by a Poisson distribution and the time between the occurrence of each event follows an exponential distribution. 

\begin{equation*}
P(X) = \frac{e^{- \lambda t} (\lambda t)^x}{x!}
\label{eq:5} \tag{5}
\end{equation*}

where $t$ is time, $E[X] = \lambda t$, and $\text{var}(X) = \lambda t$. For arrivals within the time interval $[s, t]$, $t$ is replaced with $t-s$.

### Example

*You arrive at a bus station at a random time. The buses at this station run every 10 minutes on average. What's your expected wait time? On average, when did the last bus leave the station?*

The bus arrival rate is $\lambda = \frac{1}{10}$. Due to the memoryless property of an exponential distribution, we know that the expected waiting time is $\frac{1}{\lambda} = 10$ minutes. This property also applies in the past. Therefore, on average, the bus arrived 10 minutes ago as well.  

Inituively, this doesn't make sense. Why is the expected wait time 10 minutes for a bus that arrived on average 10 minutes ago? Wouldn't that mean the buses run every 20 minutes? No because, when you arrive at a random time, the chance that you arrive during the long interval between buses is greater than the small intervals right after the last bus leaves and right before the next bus arrives.
